```python
import numpy as np
import matplotlib.pyplot as plt
```

Part 1!

```python
def getPosterior(sensitivity, specificity, prior):
    numerator = prior * sensitivity
    denominator = numerator + ((1 - prior) * (1 - specificity))
    return numerator / denominator
```

```python
prior = np.zeros(100)
for i in range(100):
    prior[i] = (i + 1) / 100
posterior = getPosterior(0.95, 0.9, prior)

plt.title("Posterior Probability vs. Prior Probability")
plt.plot(prior, posterior, color ="green")
plt.xlabel("Population with disease (percentage)")
plt.ylabel("Likelihood of having disease given a positive test")
```

Part 2!

```python
sensitivity = np.zeros(100)
for i in range(100):
    sensitivity[i] = (i + 1) / 100
posterior = getPosterior(sensitivity, 0.9, 0.01)

plt.title("Posterior Probability vs. Test Sensitivity")
plt.plot(sensitivity, posterior, color ="green")
plt.xlabel("Likelihood of a true positive result")
plt.ylabel("Likelihood of having disease given a positive test")
```

Part 3!

```python
specificity = np.zeros(100)
for i in range(100):
    specificity[i] = (i + 1) / 100
posterior = getPosterior(0.95, specificity, 0.01)

plt.title("Posterior Probability vs. Test Specificity")
plt.plot(sensitivity, posterior, color ="green")
plt.xlabel("Likelihood of a true negative result")
plt.ylabel("Likelihood of having disease given a positive test")
```

Bonus of Part 1

```python
# Making a usable set of 2d arrays for the 3D plots
def returnXY(big_x, big_y):
    x, y = [], []
    for i in range(len(big_y)):
        x.append(big_x)
    x = np.array(x)
    for j in big_y:
        sublist = []
        for k in big_x:
            sublist.append(j)
        sublist = np.array(sublist)
        y.append(sublist)
    y = np.array(y)
    return x, y
```

```python
prior = np.zeros(100)
for i in range(100):
    prior[i] = (i + 1) / 100
sensitivity = np.zeros(100)
for i in range(100):
    sensitivity[i] = (i + 1) / 100

x, y = returnXY(prior, sensitivity)
posterior = getPosterior(y, 0.9, x)

fig = plt.figure(figsize =(10, 10))
ax = plt.axes(projection ='3d')
ax.set_xlabel('Prior Probability')
ax.set_ylabel('Sensitivity Probability')
ax.set_zlabel('Posterior Probability')
plot_surface = ax.plot_surface(x, y, posterior, cmap = plt.get_cmap('cool'), edgecolor ='none')
fig.colorbar(plot_surface, ax = ax, shrink = 0.2, aspect = 10)
```
