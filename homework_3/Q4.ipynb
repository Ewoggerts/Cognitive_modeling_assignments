```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

Use Case: A professor noticed that students have begun showing up late to their class and wants to model how many times per semester students are late to class.

```python
# Parameters
mu_0 = 5.0 # Prior mean
sigma_0 = 2.0 # Prior STD
true_mu = 6.5 # True mean
sigma = 3.0 # Observation STD
n = 20 # Sample size
```

```python
# Finding data
randomData = np.random.normal(loc = true_mu, scale = sigma, size = n)
average = np.average(randomData)
variancePosterior = 1 / ((1 / sigma_0**2) + (n / sigma**2))
meanPosterior = variancePosterior * ((mu_0 / sigma_0**2) + (n * average / sigma**2))
STDposterior = np.sqrt(variancePosterior)
linspace = np.linspace(0, 15, 500)
prior, posterior= norm.pdf(linspace, mu_0, sigma_0), norm.pdf(linspace, meanPosterior, STDposterior)
```

```python
# Plot data
plt.figure(figsize=(9, 5))
plt.plot(linspace, prior, label = "Prior", color = "blue")
plt.plot(linspace, posterior, label = "Posterior", color = "red")
plt.title("Normal-Normal Model: Student Lateness per Semester")
plt.xlabel("Total Times Late in Semester (Mean)")
plt.ylabel("Density")
plt.legend()
```
